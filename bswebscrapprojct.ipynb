```python
import requests
from bs4 import BeautifulSoup
import pandas as pd
```

```python
url = "https://baccabucci.com/collections/bacca-bucci-sneakers"
```

```python
response = requests.get(url)
response
```

```python
soup = BeautifulSoup(response.text)
soup
```

```python
print(soup.prettify())
```

```python
images = []

for img in soup.find_all("img", class_="motion-reduce"):
    # try src first
    src = img.get("src")

    # if src not present, get from srcset
    if not src:
        srcset = img.get("srcset")
        if srcset:
            # take highest quality URL (last one)
            src = srcset.split(",")[-1].split(" ")[0]

    if src:
        if src.startswith("//"):
            src = "https:" + src

        # clean URL
        src = src.split("?")[0].split("&")[0]

        images.append(src)

images

```

```python
titles = []

for a in soup.find_all("a", class_="full-unstyled-link"):
    title = a.get_text(strip=True)
    if title:
        titles.append(title)

titles
```

```python
money = []

for span in soup.find_all("span", class_="price-item price-item--sale"):

    # 1. Extract text and clean it (Removing "Rs.", commas, and ".00")
    price_d = span.text.strip().replace("Rs.", "").replace(",", "").replace(".00", "").strip()

    # 2. Convert to integer
    if price_d.isdigit():
        price = int(price_d)
        money.append(price)

money
```

```python
ndmoney = []


for p in soup.find_all("dd", class_="price__compare"):

    # 1. Extract text and clean it (Removing "Rs.", commas, and ".00")
    price_t = p.text.strip().replace("Rs.", "").replace(",", "").replace(".00", "").strip()

    # 2. Convert to integer
    if price_t.isdigit():
        price = int(price_t)
        ndmoney.append(price)

ndmoney
```

```python
for product, picture, current_money, nodiscount in zip(titles,images,money,ndmoney):
  print("|",product,"|",picture,"|",current_money,"|",nodiscount,"|")
```

```python
import pandas as pd

min_length = min(len(titles), len(images), len(money), len(ndmoney))

df = pd.DataFrame({
    "Product Name": titles[:min_length],
    "Picture": images[:min_length],
    "Cost": money[:min_length],
    "OG Cost": ndmoney[:min_length]
})
df
```

```python
df.to_csv("itvbaccabucciproject.csv",index = False)
display(df)
```
